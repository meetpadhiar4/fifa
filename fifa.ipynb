```python
def weight_converter(x):
    res = x.replace('lbs', '')
    return int(res)
```

```python
fifa['Weight'] = fifa['Weight'].apply(lambda x : weight_converter(x))
```

```python
fifa['Weight']
```

```python
plt.figure(figsize=(15,8))
sns.distplot(fifa['Weight'], bins = 20, kde = False)
```

```python
def wage_converter(x):
    res = x.replace('€', '')
    if 'M' in res:
        res = int(res.replace('M', '')) * 1000000
    elif 'K' in res:
        res = int(res.replace('K', '')) * 1000
    return int(res)
    
```

```python
fifa['Wage'] = fifa['Wage'].apply(lambda x : wage_converter(x))
```

```python
plt.figure(figsize=(15, 8))
displot = sns.distplot(fifa['Wage'], bins = 100, kde = False)
displot.set_xlabel(xlabel = 'Wage')
displot.set_ylabel(ylabel = 'Number of players')
plt.show()
```

```python
players = fifa[['Name','Finishing', 'ShortPassing', 'Dribbling', 'LongPassing', 'BallControl', 'Agility', 'Vision','StandingTackle', 'Strength', 'GKReflexes']]
```

```python
cols = ['Finishing', 'ShortPassing', 'Dribbling', 'LongPassing', 'BallControl', 'Agility', 'Vision', 'StandingTackle', 'Strength', 'GKReflexes']
```

```python
def reportCard(id):
    val = list(players[cols].loc[id])
    plt.figure(figsize=(15, 8))
    plt.subplot(polar = True)
    thetha = np.linspace(0, 1.8 * 3.14, len(val))
    lines, labels = plt.thetagrids(range(0, 360, int(360 / len(val))), (cols))
    plt.plot(thetha, val)
    plt.fill(thetha, val, 'b', alpha = 0.1)
    plt.title('Report card for '+ players['Name'].loc[id])
    plt.show()
    
```

```python
reportCard(19)
```
